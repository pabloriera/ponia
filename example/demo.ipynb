### Example

Using the wage1985 dataset and asking questions in natural language.

```python
import pandas as pd
from ponia import ponia

# Cargar el dataset wage1985 con url
url = "https://raw.githubusercontent.com/stdlib-js/datasets-berndt-cps-wages-1985/refs/heads/main/data/data.csv"
df = pd.read_csv(url)
```

```python
ponia(df,"¿Cuál es el wage promedio de los trabajadores con occupation professional?", verbose=True)
```

```python
df[df['occupation'] == 'professional']['wage'].mean()
```
